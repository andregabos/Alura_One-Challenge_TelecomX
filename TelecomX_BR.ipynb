#üìå Extrac√£o

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)

df.head()
```

```python
df.info()
```

#üîß Transforma√ß√£o

```python
df_tratado = df.copy()

# Para cada coluna aninhada, normaliza e concatena ao novo DataFrame
for col in ['customer', 'phone', 'internet', 'account']:
    col_normalizada = pd.json_normalize(df_tratado[col])
    df_tratado = pd.concat([df_tratado.drop(columns=[col]), col_normalizada], axis=1)

# Faz o mesmo para a coluna 'Charges' que tamb√©m veio como um dicion√°rio dentro de 'account'
if 'Charges' in df_tratado.columns and isinstance(df_tratado['Charges'][0], dict):
    charges_normalizado = pd.json_normalize(df_tratado['Charges'])
    df_tratado = pd.concat([df_tratado.drop(columns=['Charges']), charges_normalizado], axis=1)

df_tratado.head()
```

```python
df_tratado.info()
```

```python
df_tratado.shape
```

```python
# Verifica√ß√£o de valores ausentes (NaN)
df_tratado.isnull().sum()
```

```python
# Verifica√ß√£o de dados duplicados
df_tratado.duplicated().sum()
```

```python
# Verificar se h√° IDs de cliente duplicados
df_tratado['customerID'].duplicated().sum()
```

```python
# Verifica√ß√£o de inconsist√™ncias em dados categ√≥ricos
colunas_categoricas = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'PaymentMethod', 'Churn']
for coluna in colunas_categoricas:
    if coluna in df_tratado.columns:
        print(f"Valores √∫nicos em '{coluna}': {df_tratado[coluna].unique()}")
```

```python
# Corrigindo o tipo da coluna 'Total'
df_tratado['Charges.Total'] = pd.to_numeric(df_tratado['Charges.Total'], errors='coerce')
```

```python
df_tratado['Charges.Total'].isnull().sum()
```

```python
# Preenche os valores NaN na coluna Charges.Total com 0
df_tratado['Charges.Total'].fillna(0, inplace=True)
```

```python
df_tratado['Charges.Total'].isnull().sum()
```

```python
# Padronizando os nomes das colunas para letras min√∫sculas.
df_tratado.columns = [col.lower() for col in df_tratado.columns]
```

```python
# Padronizando colunas bin√°rias ('Yes'/'No') para formato num√©rico (1/0)
colunas_binarias = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']
for coluna in colunas_binarias:
    if coluna in df_tratado.columns:
        df_tratado[coluna] = df_tratado[coluna].replace({'Yes': 1, 'No': 0})
        if coluna == 'churn':
            df_tratado[coluna] = df_tratado[coluna].replace({'': 0})
        df_tratado[coluna] = df_tratado[coluna].astype(int)
```

```python
df_tratado.info()
```

```python
# Traduzindo os nomes das colunas
mapa_colunas = {
    'customerid': 'id_cliente', 'churn': 'churn', 'gender': 'genero', 'seniorcitizen': 'idoso',
    'partner': 'parceiro', 'dependents': 'dependentes', 'tenure': 'meses_contrato',
    'phoneservice': 'servico_telefone', 'multiplelines': 'multiplas_linhas',
    'internetservice': 'servico_internet', 'onlinesecurity': 'seguranca_online',
    'onlinebackup': 'backup_online', 'deviceprotection': 'protecao_dispositivo',
    'techsupport': 'suporte_tecnico', 'streamingtv': 'streaming_tv',
    'streamingmovies': 'streaming_filmes', 'contract': 'contrato',
    'paperlessbilling': 'fatura_digital', 'paymentmethod': 'metodo_pagamento',
    'charges.monthly': 'taxa_mensal', 'charges.total': 'taxa_total'
}
df_tratado.rename(columns=mapa_colunas, inplace=True)
```

```python
# Traduzindo os valores
df_tratado['genero'] = df_tratado['genero'].replace({'Male': 'Masculino', 'Female': 'Feminino'})
df_tratado['multiplas_linhas'] = df_tratado['multiplas_linhas'].replace({'No phone service': 'Sem servi√ßo telef√¥nico', 'Yes': 'Sim', 'No': 'N√£o'})
df_tratado['servico_internet'] = df_tratado['servico_internet'].replace({'Fiber optic': 'Fibra √≥tica', 'No': 'N√£o'})

colunas_sem_internet = [
    'seguranca_online', 'backup_online', 'protecao_dispositivo',
    'suporte_tecnico', 'streaming_tv', 'streaming_filmes'
]

for col in colunas_sem_internet:
    df_tratado[col] = df_tratado[col].replace({'No internet service': 'Sem servi√ßo de internet', 'Yes': 'Sim', 'No': 'N√£o'})

df_tratado['contrato'] = df_tratado['contrato'].replace({
    'Month-to-month': 'Mensal', 'One year': 'Um ano', 'Two year': 'Dois anos'
})

df_tratado['metodo_pagamento'] = df_tratado['metodo_pagamento'].replace({
    'Electronic check': 'Cheque eletr√¥nico', 'Mailed check': 'Cheque por correio',
    'Bank transfer (automatic)': 'Transfer√™ncia banc√°ria (autom√°tica)',
    'Credit card (automatic)': 'Cart√£o de cr√©dito (autom√°tico)'
})
```

```python
df_tratado.head(10)
```

```python
# Criando a coluna 'taxa_diaria' a partir da 'taxa_mensal'
df_tratado['taxa_diaria'] = df_tratado['taxa_mensal'] / 30
df_tratado.head()
```

#üìä Carga e an√°lise

```python
# Estat√≠sticas descritivas para colunas num√©ricas
df_numericas = df_tratado[['meses_contrato', 'taxa_mensal', 'taxa_total', 'taxa_diaria']]
df_numericas.describe().round(2)
```

```python
# Estat√≠sticas descritivas para colunas categ√≥ricas (texto)
df_categoricas = df_tratado.drop(columns=['meses_contrato', 'taxa_mensal', 'taxa_total', 'taxa_diaria'])
df_categoricas.describe(include='all')
```

```python
import plotly.express as px

colunas_categoricas_plot = df_tratado.select_dtypes(include='object').columns.tolist()
if 'id_cliente' in colunas_categoricas_plot:
    colunas_categoricas_plot.remove('id_cliente')

colunas_binarias_plot = ['idoso' ,'parceiro', 'dependentes', 'servico_telefone', 'fatura_digital']

for i, col in enumerate(colunas_categoricas_plot):
    fig = px.histogram(
        df_tratado,
        x=col,
        color='churn',
        barmode='group'
    )

    if col in colunas_binarias_plot:
        fig.update_xaxes(
            tickmode='array',
            tickvals=[0, 1],
            ticktext=['N√£o', 'Sim']
        )

    fig.update_yaxes(title_text='N√∫mero de Clientes')

    fig.update_layout(
        title_text=f'Distribui√ß√£o de {col.replace("_", " ").title()} por Status de Evas√£o',
        showlegend=True,
        legend_title='Evas√£o'
    )

    fig.for_each_trace(lambda t: t.update(name = 'N√£o Evadiu' if t.name == '0' else 'Evadiu'))

    fig.show()
```

```python
import plotly.express as px

for col in colunas_categoricas_plot:
    # Calculo de propor√ß√£o
    churn_proportions = df_tratado.groupby(col)['churn'].value_counts(normalize=True).mul(100).reset_index(name='porcentagem')

    fig = px.bar(
        churn_proportions,
        x=col,
        y='porcentagem',
        color=churn_proportions['churn'].astype(str), # Trata 'churn' como string para colora√ß√£o correta
        text='porcentagem',
        barmode='stack'
    )

    fig.update_layout(
        title=f'Propor√ß√£o de Evas√£o por {col.replace("_", " ").title()}',
        xaxis_title=col.replace("_", " ").title(),
        yaxis_title='Porcentagem',
        legend_title='Evas√£o',
    )

    fig.update_traces(texttemplate='%{text:.1f}%', textposition='inside')

    fig.for_each_trace(lambda t: t.update(name = 'N√£o Evadiu' if t.name == '0' else 'Evadiu'))


    fig.show()
```

```python
fig = px.histogram(
    df_tratado,
    x='meses_contrato',
    text_auto=True,
    color='churn',
    barmode='group'
)

fig.update_layout(
    title='Distribui√ß√£o de Clientes por Tempo de Contrato e Evas√£o',
    xaxis_title='Meses de Contrato',
    yaxis_title='N√∫mero de Clientes',
    legend_title='Evas√£o'
)

fig.for_each_trace(lambda t: t.update(name='N√£o Evadiu' if t.name == '0' else 'Evadiu'))

fig.show()
```

```python
df_agrupado = df_tratado.groupby(['meses_contrato', 'churn']).size().unstack(fill_value=0)

df_agrupado['churn_rate'] = (df_agrupado.get(1, 0) / df_agrupado.sum(axis=1)) * 100

fig = px.line(df_agrupado.reset_index(),
              x='meses_contrato',
              y='churn_rate',
              markers=True,
              title='Taxa de Evas√£o por Tempo de Contrato')

fig.update_layout(
    title='Distribui√ß√£o de Clientes por Tempo de Contrato e Evas√£o',
    xaxis_title='Meses de Contrato',
    yaxis_title='Taxa de Evas√£o',
    showlegend=False
)

fig.show()
```

# Relat√≥rio de An√°lise de Evas√£o de Clientes (Churn)

## 1. Introdu√ß√£o
A evas√£o de clientes, tamb√©m conhecida como churn, representa um desafio estrat√©gico para empresas do setor de telecomunica√ß√µes, impactando diretamente a receita recorrente, a lucratividade e o crescimento sustent√°vel do neg√≥cio.

A compreens√£o dos fatores que levam os clientes a encerrar seus contratos √© essencial para o desenvolvimento de estrat√©gias de reten√ß√£o mais eficazes. Identificar padr√µes de comportamento e caracter√≠sticas comuns entre os clientes que evadem permite a√ß√µes proativas para mitigar esse fen√¥meno.

Este relat√≥rio tem como objetivo analisar o conjunto de dados fornecido pela empresa fict√≠cia TelecomX, o qual cont√©m informa√ß√µes detalhadas sobre o perfil, servi√ßos contratados e hist√≥rico de faturamento de cada cliente. Atrav√©s da an√°lise explorat√≥ria dos dados, buscamos identificar os principais fatores associados ao churn e gerar insights que subsidiem a tomada de decis√µes estrat√©gicas.

A an√°lise ser√° conduzida em tr√™s etapas principais:

1. Pr√©-processamento dos dados: normaliza√ß√£o de colunas aninhadas, tratamento de valores ausentes e padroniza√ß√£o de vari√°veis.

2.	An√°lise explorat√≥ria (EDA): investiga√ß√£o visual e estat√≠stica de vari√°veis demogr√°ficas, contratuais e de consumo, com foco na compara√ß√£o entre clientes que evadiram e os que permaneceram.

3.	Gera√ß√£o de recomenda√ß√µes: proposi√ß√£o de a√ß√µes baseadas nos padr√µes observados, visando a redu√ß√£o da taxa de churn.


### Descri√ß√£o dos Dados
O dataset fornecido pela TelecomX √© estruturado em formato JSON e cont√©m registros aninhados para cada cliente. Cada entrada representa um cliente √∫nico identificado por customerID e cont√©m as seguintes se√ß√µes:
- Churn: Indica se o cliente deixou (Yes) ou permaneceu (No) com a empresa.
- customer: Informa√ß√µes demogr√°ficas e de relacionamento, como:
  - gender: G√™nero
  - SeniorCitizen: Indicador bin√°rio de idade (1 para idoso)
  - Partner e Dependents: Presen√ßa de parceiro e dependentes
  - tenure: Meses de contrato
- phone: Servi√ßos de telefonia:
  - PhoneService e MultipleLines
- internet: Servi√ßos de internet e adicionais:
  - InternetService, OnlineSecurity, OnlineBackup, DeviceProtection
  - TechSupport, StreamingTV, StreamingMovies
- account: Informa√ß√µes contratuais e de faturamento:
  - Contract, PaperlessBilling, PaymentMethod
- Charges: valores de Monthly (mensal) e Total (total acumulado)

O dataset requer pr√©-processamento para normalizar as colunas aninhadas e converter dados categ√≥ricos em formatos mais apropriados para an√°lise.

## 2. Limpeza e Tratamento de Dados

A etapa de limpeza e tratamento foi essencial para transformar o conjunto de dados original, com estrutura aninhada e diversos valores categ√≥ricos, em um formato plano, limpo e apropriado para an√°lise estat√≠stica e modelagem. As principais a√ß√µes realizadas foram as seguintes:

#### 2.1 Carregamento dos Dados

Os dados foram carregados a partir de um arquivo JSON hospedado em um reposit√≥rio GitHub utilizando a biblioteca *pandas*, com a fun√ß√£o `pd.read_json()`.

#### 2.2 Normaliza√ß√£o de Colunas Aninhadas

As colunas customer, phone, internet e account apresentavam estruturas do tipo dicion√°rio (aninhadas). Para facilitar a an√°lise, essas colunas foram normalizadas utilizando `pd.json_normalize()` com separador de underscore (_).

#### 2.3 Verifica√ß√£o de Valores Ausentes e Duplicados

Foi realizada a contagem de valores nulos utilizando o m√©todo `.isnull().sum()` e de linhas duplicadas utilizando o m√©todo `.duplicated().sum()` para garantir a consist√™ncia dos dados. Identificou-se que a coluna 'Charges.Total' possu√≠a 11 valores ausentes ap√≥s a normaliza√ß√£o e convers√£o para num√©rico, enquanto n√£o foram encontradas linhas duplicadas ou IDs de cliente duplicados.

#### 2.4 Tratamento de Valores Ausentes

Ap√≥s convers√£o de Charges.Total para num√©rico, usando o m√©todo `.to_numeric()` com o par√¢metro `errors=coerce`, para transformar os valores vazios em *NaN*, 11 registros apresentaram valores nulos. Esses valores foram preenchidos com zero, assumindo que correspondem a clientes com faturamento ainda n√£o iniciado. Para isso foi aplicado `.fillna(0, inplace=True)`.


#### 2.5 Padroniza√ß√£o de Nomes de Colunas

Para padroniza√ß√£o e legibilidade, todos os nomes de colunas foram convertidos para letras min√∫sculas usando o m√©todo `.lower()` junto de um *for loop*.


#### 2.6 Convers√£o de Vari√°veis Categ√≥ricas Bin√°rias

Colunas com respostas ‚ÄúYes‚Äù/‚ÄúNo‚Äù foram convertidas para vari√°veis bin√°rias (1/0). Valores vazios tamb√©m foram tratados. Para isso foi utilizado o m√©todo `.replace()` para substitui√ß√£o dos valores, o m√©todo `.astype(int)` para ajuste do tipo, junto de um *for loop* para altera√ß√£o de todas as colunas.


#### 2.7 Tradu√ß√£o de Colunas e Valores

Para facilitar a leitura e interpreta√ß√£o, as colunas e valores foram traduzidos para o portugu√™s com base em um dicion√°rio de mapeamento e usando o m√©todo `.replace()` para substituir os valores.

#### 2.8 Cria√ß√£o da Coluna taxa_diaria

Foi criada a vari√°vel taxa_diaria, uma estimativa do custo di√°rio com base na taxa_mensal, dividida por 30.


#### 2.9 Resultado Final da Limpeza

Ao final do processo, os dados foram armazenados no DataFrame df_tratado, com estrutura tabular plana e colunas preparadas para an√°lise.

### Distribui√ß√£o de Vari√°veis Categ√≥ricas por Churn

Para complementar os histogramas, foram criados gr√°ficos de barras empilhadas mostrando a propor√ß√£o de churn dentro de cada categoria. Esses gr√°ficos refor√ßaram as observa√ß√µes dos histogramas, destacando as categorias com as maiores taxas de evas√£o, como clientes com Fibra √ìtica, sem servi√ßos de seguran√ßa/suporte online, com fatura digital e que pagam via cheque eletr√¥nico.

```python
# @title
for col in colunas_categoricas_plot:
    # Calculo de propor√ß√£o
    churn_proportions = df_tratado.groupby(col)['churn'].value_counts(normalize=True).mul(100).reset_index(name='percentage')

    fig = px.bar(
        churn_proportions,
        x=col,
        y='percentage',
        color=churn_proportions['churn'].astype(str),
        text='percentage',
        barmode='stack'
    )

    fig.update_layout(
        title=f'Propor√ß√£o de Churn por {col.replace("_", " ").title()}',
        xaxis_title=col.replace("_", " ").title(),
        yaxis_title='Porcentagem',
        legend_title='Evas√£o',
        width=1000,
        height=400
    )

    fig.update_traces(texttemplate='%{text:.1f}%', textposition='inside')

    fig.for_each_trace(lambda t: t.update(name = 'N√£o Evadiu' if t.name == '0' else 'Evadiu'))


    fig.show()
```

*   **Parceiro e Dependentes:** Clientes sem parceiros e/ou dependentes parecem evadir mais.

*   **Servi√ßo de Internet:** Clientes com servi√ßo de internet de Fibra √ìtica apresentam uma taxa de evas√£o visivelmente maior do que aqueles com DSL ou sem servi√ßo de internet.
*   **Servi√ßos Online (Seguran√ßa, Backup, Prote√ß√£o, Suporte):** Clientes que *n√£o* possuem servi√ßos adicionais online (Seguran√ßa Online, Backup Online, Prote√ß√£o de Dispositivo, Suporte T√©cnico) tendem a evadir mais.
*   **Streaming (TV e Filmes):** Clientes que utilizam servi√ßos de streaming (TV e Filmes) parecem ter uma taxa de evas√£o ligeiramente menor.
*   **Fatura Digital:** Clientes que optam por fatura digital parecem ter uma taxa de evas√£o maior.
*   **M√©todo de Pagamento:** O m√©todo de pagamento "Cheque eletr√¥nico" se destaca com uma taxa de evas√£o consideravelmente mais alta em compara√ß√£o com outros m√©todos.

### Taxa de Churn por Tempo de Contrato

Um gr√°fico de linha mostrando a taxa de churn (%) por meses de contrato (`meses_contrato`) revelou que a taxa de evas√£o √© extremamente alta nos primeiros meses de contrato e diminui drasticamente √† medida que o tempo de contrato aumenta.
