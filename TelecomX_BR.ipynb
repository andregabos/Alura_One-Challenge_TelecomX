#📌 Extracão

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)

df.head()
```

```python
df.info()
```

#🔧 Transformação

```python
df_tratado = df.copy()

# Para cada coluna aninhada, normaliza e concatena ao novo DataFrame
for col in ['customer', 'phone', 'internet', 'account']:
    col_normalizada = pd.json_normalize(df_tratado[col])
    df_tratado = pd.concat([df_tratado.drop(columns=[col]), col_normalizada], axis=1)

# Faz o mesmo para a coluna 'Charges' que também veio como um dicionário dentro de 'account'
if 'Charges' in df_tratado.columns and isinstance(df_tratado['Charges'][0], dict):
    charges_normalizado = pd.json_normalize(df_tratado['Charges'])
    df_tratado = pd.concat([df_tratado.drop(columns=['Charges']), charges_normalizado], axis=1)

df_tratado.head()
```

```python
df_tratado.info()
```

```python
df_tratado.shape
```

```python
# Verificação de valores ausentes (NaN)
df_tratado.isnull().sum()
```

```python
# Verificação de dados duplicados
df_tratado.duplicated().sum()
```

```python
# Verificar se há IDs de cliente duplicados
df_tratado['customerID'].duplicated().sum()
```

```python
# Verificação de inconsistências em dados categóricos
colunas_categoricas = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'PaymentMethod', 'Churn']
for coluna in colunas_categoricas:
    if coluna in df_tratado.columns:
        print(f"Valores únicos em '{coluna}': {df_tratado[coluna].unique()}")
```

```python
# Corrigindo o tipo da coluna 'Total'
df_tratado['Charges.Total'] = pd.to_numeric(df_tratado['Charges.Total'], errors='coerce')
```

```python
df_tratado['Charges.Total'].isnull().sum()
```

```python
# Preenche os valores NaN na coluna Charges.Total com 0
df_tratado['Charges.Total'].fillna(0, inplace=True)
```

```python
df_tratado['Charges.Total'].isnull().sum()
```

```python
# Padronizando os nomes das colunas para letras minúsculas.
df_tratado.columns = [col.lower() for col in df_tratado.columns]
```

```python
# Padronizando colunas binárias ('Yes'/'No') para formato numérico (1/0)
colunas_binarias = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']
for coluna in colunas_binarias:
    if coluna in df_tratado.columns:
        df_tratado[coluna] = df_tratado[coluna].replace({'Yes': 1, 'No': 0})
        if coluna == 'churn':
            df_tratado[coluna] = df_tratado[coluna].replace({'': 0})
        df_tratado[coluna] = df_tratado[coluna].astype(int)
```

```python
df_tratado.info()
```

```python
# Traduzindo os nomes das colunas
mapa_colunas = {
    'customerid': 'id_cliente', 'churn': 'churn', 'gender': 'genero', 'seniorcitizen': 'idoso',
    'partner': 'parceiro', 'dependents': 'dependentes', 'tenure': 'meses_contrato',
    'phoneservice': 'servico_telefone', 'multiplelines': 'multiplas_linhas',
    'internetservice': 'servico_internet', 'onlinesecurity': 'seguranca_online',
    'onlinebackup': 'backup_online', 'deviceprotection': 'protecao_dispositivo',
    'techsupport': 'suporte_tecnico', 'streamingtv': 'streaming_tv',
    'streamingmovies': 'streaming_filmes', 'contract': 'contrato',
    'paperlessbilling': 'fatura_digital', 'paymentmethod': 'metodo_pagamento',
    'charges.monthly': 'taxa_mensal', 'charges.total': 'taxa_total'
}
df_tratado.rename(columns=mapa_colunas, inplace=True)
```

```python
# Traduzindo os valores
df_tratado['genero'] = df_tratado['genero'].replace({'Male': 'Masculino', 'Female': 'Feminino'})
df_tratado['multiplas_linhas'] = df_tratado['multiplas_linhas'].replace({'No phone service': 'Sem serviço telefônico', 'Yes': 'Sim', 'No': 'Não'})
df_tratado['servico_internet'] = df_tratado['servico_internet'].replace({'Fiber optic': 'Fibra ótica', 'No': 'Não'})

colunas_sem_internet = [
    'seguranca_online', 'backup_online', 'protecao_dispositivo',
    'suporte_tecnico', 'streaming_tv', 'streaming_filmes'
]

for col in colunas_sem_internet:
    df_tratado[col] = df_tratado[col].replace({'No internet service': 'Sem serviço de internet', 'Yes': 'Sim', 'No': 'Não'})

df_tratado['contrato'] = df_tratado['contrato'].replace({
    'Month-to-month': 'Mensal', 'One year': 'Um ano', 'Two year': 'Dois anos'
})

df_tratado['metodo_pagamento'] = df_tratado['metodo_pagamento'].replace({
    'Electronic check': 'Cheque eletrônico', 'Mailed check': 'Cheque por correio',
    'Bank transfer (automatic)': 'Transferência bancária (automática)',
    'Credit card (automatic)': 'Cartão de crédito (automático)'
})
```

```python
df_tratado.head(10)
```

```python
# Criando a coluna 'taxa_diaria' a partir da 'taxa_mensal'
df_tratado['taxa_diaria'] = df_tratado['taxa_mensal'] / 30
df_tratado.head()
```

```python
import plotly.express as px

for col in colunas_categoricas_plot:
    # Calculo de proporção
    churn_proportions = df_tratado.groupby(col)['churn'].value_counts(normalize=True).mul(100).reset_index(name='porcentagem')

    fig = px.bar(
        churn_proportions,
        x=col,
        y='porcentagem',
        color=churn_proportions['churn'].astype(str), # Trata 'churn' como string para coloração correta
        text='porcentagem',
        barmode='stack'
    )

    fig.update_layout(
        title=f'Proporção de Evasão por {col.replace("_", " ").title()}',
        xaxis_title=col.replace("_", " ").title(),
        yaxis_title='Porcentagem',
        legend_title='Evasão',
    )

    fig.update_traces(texttemplate='%{text:.1f}%', textposition='inside')

    fig.for_each_trace(lambda t: t.update(name = 'Não Evadiu' if t.name == '0' else 'Evadiu'))


    fig.show()
```

```python
fig = px.histogram(
    df_tratado,
    x='meses_contrato',
    text_auto=True,
    color='churn',
    barmode='group'
)

fig.update_layout(
    title='Distribuição de Clientes por Tempo de Contrato e Evasão',
    xaxis_title='Meses de Contrato',
    yaxis_title='Número de Clientes',
    legend_title='Evasão'
)

fig.for_each_trace(lambda t: t.update(name='Não Evadiu' if t.name == '0' else 'Evadiu'))

fig.show()
```

```python
df_agrupado = df_tratado.groupby(['meses_contrato', 'churn']).size().unstack(fill_value=0)

df_agrupado['churn_rate'] = (df_agrupado.get(1, 0) / df_agrupado.sum(axis=1)) * 100

fig = px.line(df_agrupado.reset_index(),
              x='meses_contrato',
              y='churn_rate',
              markers=True,
              title='Taxa de Evasão por Tempo de Contrato')

fig.update_layout(
    title='Distribuição de Clientes por Tempo de Contrato e Evasão',
    xaxis_title='Meses de Contrato',
    yaxis_title='Taxa de Evasão',
    showlegend=False
)

fig.show()
```

# Relatório de Análise de Evasão de Clientes (Churn)

## 1. Introdução
A evasão de clientes, também conhecida como churn, representa um desafio estratégico para empresas do setor de telecomunicações, impactando diretamente a receita recorrente, a lucratividade e o crescimento sustentável do negócio.

A compreensão dos fatores que levam os clientes a encerrar seus contratos é essencial para o desenvolvimento de estratégias de retenção mais eficazes. Identificar padrões de comportamento e características comuns entre os clientes que evadem permite ações proativas para mitigar esse fenômeno.

Este relatório tem como objetivo analisar o conjunto de dados fornecido pela empresa fictícia TelecomX, o qual contém informações detalhadas sobre o perfil, serviços contratados e histórico de faturamento de cada cliente. Através da análise exploratória dos dados, buscamos identificar os principais fatores associados ao churn e gerar insights que subsidiem a tomada de decisões estratégicas.

A análise será conduzida em três etapas principais:

1. Pré-processamento dos dados: normalização de colunas aninhadas, tratamento de valores ausentes e padronização de variáveis.

2.	Análise exploratória (EDA): investigação visual e estatística de variáveis demográficas, contratuais e de consumo, com foco na comparação entre clientes que evadiram e os que permaneceram.

3.	Geração de recomendações: proposição de ações baseadas nos padrões observados, visando a redução da taxa de churn.


### Descrição dos Dados
O dataset fornecido pela TelecomX é estruturado em formato JSON e contém registros aninhados para cada cliente. Cada entrada representa um cliente único identificado por customerID e contém as seguintes seções:
- Churn: Indica se o cliente deixou (Yes) ou permaneceu (No) com a empresa.
- customer: Informações demográficas e de relacionamento, como:
  - gender: Gênero
  - SeniorCitizen: Indicador binário de idade (1 para idoso)
  - Partner e Dependents: Presença de parceiro e dependentes
  - tenure: Meses de contrato
- phone: Serviços de telefonia:
  - PhoneService e MultipleLines
- internet: Serviços de internet e adicionais:
  - InternetService, OnlineSecurity, OnlineBackup, DeviceProtection
  - TechSupport, StreamingTV, StreamingMovies
- account: Informações contratuais e de faturamento:
  - Contract, PaperlessBilling, PaymentMethod
- Charges: valores de Monthly (mensal) e Total (total acumulado)

O dataset requer pré-processamento para normalizar as colunas aninhadas e converter dados categóricos em formatos mais apropriados para análise.

## 2. Limpeza e Tratamento de Dados

A etapa de limpeza e tratamento foi essencial para transformar o conjunto de dados original, com estrutura aninhada e diversos valores categóricos, em um formato plano, limpo e apropriado para análise estatística e modelagem. As principais ações realizadas foram as seguintes:

#### 2.1 Carregamento dos Dados

Os dados foram carregados a partir de um arquivo JSON hospedado em um repositório GitHub utilizando a biblioteca *pandas*, com a função `pd.read_json()`.

#### 2.2 Normalização de Colunas Aninhadas

As colunas customer, phone, internet e account apresentavam estruturas do tipo dicionário (aninhadas). Para facilitar a análise, essas colunas foram normalizadas utilizando `pd.json_normalize()` com separador de underscore (_).

#### 2.3 Verificação de Valores Ausentes e Duplicados

Foi realizada a contagem de valores nulos utilizando o método `.isnull().sum()` e de linhas duplicadas utilizando o método `.duplicated().sum()` para garantir a consistência dos dados. Identificou-se que a coluna 'Charges.Total' possuía 11 valores ausentes após a normalização e conversão para numérico, enquanto não foram encontradas linhas duplicadas ou IDs de cliente duplicados.

#### 2.4 Tratamento de Valores Ausentes

Após conversão de Charges.Total para numérico, usando o método `.to_numeric()` com o parâmetro `errors=coerce`, para transformar os valores vazios em *NaN*, 11 registros apresentaram valores nulos. Esses valores foram preenchidos com zero, assumindo que correspondem a clientes com faturamento ainda não iniciado. Para isso foi aplicado `.fillna(0, inplace=True)`.


#### 2.5 Padronização de Nomes de Colunas

Para padronização e legibilidade, todos os nomes de colunas foram convertidos para letras minúsculas usando o método `.lower()` junto de um *for loop*.


#### 2.6 Conversão de Variáveis Categóricas Binárias

Colunas com respostas “Yes”/“No” foram convertidas para variáveis binárias (1/0). Valores vazios também foram tratados. Para isso foi utilizado o método `.replace()` para substituição dos valores, o método `.astype(int)` para ajuste do tipo, junto de um *for loop* para alteração de todas as colunas.


#### 2.7 Tradução de Colunas e Valores

Para facilitar a leitura e interpretação, as colunas e valores foram traduzidos para o português com base em um dicionário de mapeamento e usando o método `.replace()` para substituir os valores.

#### 2.8 Criação da Coluna taxa_diaria

Foi criada a variável taxa_diaria, uma estimativa do custo diário com base na taxa_mensal, dividida por 30.


#### 2.9 Resultado Final da Limpeza

Ao final do processo, os dados foram armazenados no DataFrame df_tratado, com estrutura tabular plana e colunas preparadas para análise.

## Conclusões e Insights

A análise exploratória dos dados revelou insights importantes sobre os fatores que parecem estar mais associados à evasão de clientes na TelecomX. Um dos achados mais significativos diz respeito ao **tempo de contrato**, que se mostrou um fator crítico. A taxa de churn é significativamente maior nos primeiros meses de contrato, diminuindo consideravelmente à medida que o cliente permanece mais tempo com a empresa. Isso sugere que os primeiros meses representam um período de alto risco para a perda de clientes, e a retenção nesse estágio inicial é fundamental.

Os **serviços de internet e adicionais** também desempenham um papel importante na evasão. Clientes com serviço de internet de **Fibra Ótica** apresentaram uma propensão maior à evasão em comparação com aqueles que utilizam DSL ou que não possuem serviço de internet. Adicionalmente, a ausência de serviços adicionais online, como **Segurança Online, Backup Online, Proteção de Dispositivo e Suporte Técnico**, está associada a taxas de churn mais altas. Este padrão pode indicar insatisfação com a qualidade percebida do serviço de Fibra Ótica ou uma falta de valor atribuído aos serviços de segurança e suporte, levando os clientes a buscar alternativas.

O **método de pagamento** utilizado pelo cliente também se destaca como um fator relevante. O método de pagamento "Cheque eletrônico" apresentou uma taxa de evasão notavelmente superior aos outros métodos. Isso pode estar relacionado a potenciais problemas na experiência de pagamento associada a este método ou a um perfil de cliente que prefere essa forma de pagamento e que, por sua vez, demonstra maior propensão a descontinuar o serviço.

Algumas **características do cliente** também foram identificadas como potencialmente associadas à evasão. Clientes considerados **idosos**, sem **parceiros** e sem **dependentes** parecem evadir mais frequentemente. Embora essas características demográficas não sejam diretamente controláveis pela empresa, elas podem ser utilizadas para identificar segmentos de clientes de maior risco e direcionar esforços de retenção de forma mais eficaz.

A análise das **taxas mensais e totais** também forneceu insights relevantes. Clientes que evadiram tendem a ter **taxas mensais** mais altas e **taxas totais** menores, o que é consistente com um tempo de contrato mais curto. Isso sugere que o custo do serviço em relação ao tempo de permanência pode ser um fator decisivo para o churn, particularmente para clientes com planos de maior custo e menor tempo de relacionamento com a empresa.

Por fim, a opção pela **fatura digital** também parece estar associada a uma taxa de evasão maior. Este achado pode estar correlacionado com outros fatores ou pode indicar uma preferência por menos interação direta com a empresa, o que, em alguns casos, pode preceder a decisão de cancelar o serviço.

Em suma, os principais impulsionadores de churn identificados nesta análise parecem ser o **curto tempo de contrato**, a contratação de **serviço de internet de Fibra Ótica** sem serviços de segurança/suporte online, o uso do **método de pagamento Cheque eletrônico**, e **taxas mensais mais altas** para clientes com menor tempo de contrato. Estes insights fornecem uma base sólida para o desenvolvimento de estratégias de retenção direcionadas e eficazes.

## Recomendações

Com base nos insights derivados da análise exploratória, as seguintes recomendações são propostas para a TelecomX com o objetivo de reduzir a evasão de clientes:

**Foco na Integração e Experiência Inicial:** Dado que a taxa de churn é mais alta nos primeiros meses, é crucial investir em um processo de integração de novos clientes robusto. Isso pode incluir comunicação proativa com novos clientes para garantir que a instalação e a configuração dos serviços ocorreram sem problemas, oferecimento de canais de suporte de fácil acesso e com tempo de resposta rápido, disponibilização de conteúdo educativo com tutoriais e guias claros sobre como usar os serviços e aproveitar ao máximo o plano contratado, e a realização de pesquisas de satisfação iniciais para coletar feedback nos primeiros meses, visando identificar e resolver problemas antes que levem ao churn.

**Avaliar a Experiência com Fibra Ótica e Promover Serviços Adicionais:** A alta taxa de churn entre usuários de Fibra Ótica, especialmente aqueles sem serviços de segurança/suporte online, sugere possíveis problemas de qualidade percebida ou falta de valor dos serviços adicionais. Recomenda-se investigar a fundo a satisfação de clientes de Fibra Ótica através de pesquisas direcionadas para entender as causas da insatisfação, melhorar a comunicação sobre o valor dos serviços adicionais destacando os benefícios da segurança online, backup, proteção de dispositivo e suporte técnico (talvez oferecendo pacotes promocionais nos primeiros meses), e garantir a qualidade e estabilidade do serviço de Fibra Ótica monitorando de perto a performance da rede e resolvendo proativamente quaisquer problemas relatados.

**Otimizar a Experiência de Pagamento via Cheque Eletrônico:** A alta taxa de churn associada a este método de pagamento exige atenção. É importante investigar possíveis problemas técnicos ou de usabilidade analisando o processo de pagamento por cheque eletrônico para identificar fricções ou dificuldades, oferecer incentivos para outros métodos de pagamento encorajando os clientes a migrarem para métodos de pagamento automático mais convenientes e com menor taxa de churn (cartão de crédito ou transferência bancária automática), e melhorar a comunicação sobre o processo de pagamento garantindo que os clientes entendam claramente como funciona o pagamento por cheque eletrônico e quaisquer requisitos associados.

**Estratégias de Retenção Segmentadas:** Utilizar as características demográficas e de serviço identificadas (idosos, sem parceiros/dependentes) para criar ofertas de retenção direcionadas. Por exemplo, pacotes de serviços que atendam melhor às necessidades desses segmentos ou programas de fidelidade específicos podem ser implementados.

**Análise da Relação Preço-Valor:** Para clientes com taxas mensais mais altas e menor tempo de contrato, reavaliar a estrutura de preços e considerar a oferta de planos mais flexíveis que se ajustem melhor ao uso e orçamento do cliente, além de garantir que os clientes percebam o valor que estão recebendo pelos serviços, especialmente em planos de maior custo, através de comunicação clara do valor.

**Revisar a Experiência da Fatura Digital:** Investigar por que clientes com fatura digital evadem mais. Isso pode estar ligado a problemas no recebimento ou acesso à fatura, ou pode ser um sintoma de um perfil de cliente mais propenso a gerenciar serviços de forma menos "conectada" à empresa, o que, em alguns casos, pode preceder a decisão de cancelar o serviço.

Implementar estas recomendações de forma estratégica e monitorar seu impacto na taxa de churn será fundamental para a TelecomX melhorar a retenção de clientes e garantir um crescimento sustentável.
